```python
import pandas as pd
from IPython.display import display_html
import requests
import numpy as np
import itertools
import datetime
import locale
from dateutil.parser import parse

locale.setlocale(locale.LC_TIME, 'nl_NL')
pd.options.mode.use_inf_as_na = True
URL = f"https://rooster.horizoncollege.nl/rstr/ECO/AMR/400-ECO/Roosters/c/37/c00030.htm"
tables = pd.read_html(URL, header=0)
df = tables[0]
df = df.where(df.notnull(), 0)
df = df.rename(columns={"Unnamed: 0": "Uur"})

for a, b in itertools.combinations(df, 2):
    adata = a.split()
    bdata = b.split()
    if adata[0] == bdata[0]:
        try:
            if (df[a] == df[b]).all():
                df = df.drop(b, axis=1)
        except:
            continue

def get_day() -> str:
    now = datetime.datetime.now()
    day = now.strftime("%A").capitalize()
    date = now.strftime("%d-%m")
    return f"{day} {date}"

def is_time_between(begin_time, end_time, check_time=None):
    check_time = check_time or datetime.utcnow().time()
    if begin_time < end_time:
        return check_time >= begin_time and check_time <= end_time
    else:
        return check_time >= begin_time or check_time <= end_time
    
#df.filter(regex=get_day())
t = "13:39"
for count, v in enumerate(df["Uur"]):
    d = v.split()
    if is_time_between(parse(d[1]), parse(d[2]), parse(t)):
        rooster = df.filter(regex=get_day())
        print(rooster.iloc[count])
        break
    
```
