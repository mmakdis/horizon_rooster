```python
import pandas as pd
from IPython.display import display_html
import requests
import numpy as np
import itertools
pd.options.mode.use_inf_as_na = True
URL = f"https://rooster.horizoncollege.nl/rstr/ECO/AMR/400-ECO/Roosters/c/37/c00030.htm"
tables = pd.read_html(URL, header=0)
df = tables[0]
df = df.where(df.notnull(), 0)
df = df.rename(columns={"Unnamed: 0": "Uur"})

for a, b in itertools.combinations(df, 2):
    adata = a.split()
    bdata = b.split()
    if adata[0] == "Maandag" and bdata[0] == "Maandag":
        if (df[a] == df[b]).all():
            try:
                df = df.drop(b, axis=1)
            except:
                continue

df
```
